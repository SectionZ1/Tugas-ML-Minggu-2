### Keterangan Latihan:
Latihan ini bertujuan untuk mempraktikkan **Pipeline Lengkap Preprocessing Data** sebelum melatih model Machine Learning:
1. **Handling Missing Values** - Mengimputasi missing values numerik menggunakan `SimpleImputer` dengan strategi mean.
2. **Categorical Encoding** - Mengubah kolom kategori menjadi dummy variables biner menggunakan `ColumnTransformer` dan `OneHotEncoder`.
3. **Target Label Encoding** - Mengodekan label target biner menggunakan `LabelEncoder`.
4. **Splitting Data** - Membagi dataset menjadi training set dan test set.
5. **Feature Scaling** - Menstandarkan rentang data numerik menggunakan `StandardScaler`.

Dataset yang digunakan: `Data.csv`.

## 1. Import Library

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

## 2. Load Dataset & Pisahkan Fitur (X) dan Target (y)
Kita memuat file `Data.csv`. Kolom terakhir (`Purchased`) merupakan target label (y) dan kolom lainnya adalah fitur (X).

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

print("Fitur Awal (X):")
print(X)
print("\nTarget Awal (y):")
print(y)
```

## 3. Imputasi Nilai Kosong (Missing Values)
Kita menggunakan `SimpleImputer` dengan strategi 'mean' untuk mengisi data kosong pada kolom indeks 1 (Age) dan indeks 2 (Salary).

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])

print("Fitur setelah Imputasi Nilai Kosong (X):")
print(X)
```

## 4. Encoding Data Kategorikal pada Fitur X
Kita menggunakan `ColumnTransformer` bersama dengan `OneHotEncoder` untuk kolom indeks 0 (Country) agar diubah menjadi representasi One-Hot biner.

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

print("Fitur setelah One-Hot Encoding pada Kolom 'Country' (X):")
print(X)
```

## 5. Encoding Label Target (y)
Kita menggunakan `LabelEncoder` untuk mengubah target biner ('No' / 'Yes') menjadi representasi angka (0 / 1).

```python
le = LabelEncoder()
y = le.fit_transform(y)

print("Target setelah Label Encoding (y):")
print(y)
```

## 6. Pemisahan Dataset (Train & Test Split)
Membagi dataset menjadi 80% data training dan 20% data testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

print("Ukuran X_train:", X_train.shape)
print("Ukuran X_test:", X_test.shape)
print("X_train:")
print(X_train)
```

## 7. Feature Scaling (Standardisasi)
Menstandarkan kolom numerik (Age dan Salary yang berada pada indeks 3 dan 4 setelah proses One-Hot Encoding) menggunakan `StandardScaler`.

```python
sc = StandardScaler()
# Melakukan fit & transform pada data train, serta transform saja pada data test
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])

print("X_train setelah Feature Scaling:")
print(X_train)
print("\nX_test setelah Feature Scaling:")
print(X_test)
```

## 8. Kesimpulan
Melalui tahapan pipeline preprocessing di atas:
1. Data kosong telah diatasi menggunakan imputasi nilai rata-rata.
2. Variabel kategorik non-ordinal dikodekan dengan One-Hot Encoding.
3. Variabel target biner dikodekan dengan Label Encoding.
4. Data dipisahkan agar evaluasi model objektif.
5. Fitur numerik distandardisasi dengan StandardScaler sehingga model tidak sensitif terhadap perbedaan skala nilai yang ekstrem.
