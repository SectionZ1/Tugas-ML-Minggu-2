### Keterangan Latihan:
Latihan ini bertujuan untuk mempraktikkan berbagai teknik **Categorical Encoding** yang digunakan untuk mengubah data kategorik menjadi data numerik agar dapat diproses oleh algoritma Machine Learning:

1. **Label Encoding** – Mengubah kategori menjadi angka integer secara berurutan.
2. **Ordinal Encoding** – Mengubah kategori yang memiliki urutan menjadi nilai numerik yang mencerminkan urutan tersebut.
3. **One-Hot Encoding (manual)** – Membuat kolom biner untuk setiap kategori unik (diterapkan secara manual tanpa library khusus).

Dataset yang digunakan: data buatan sendiri yang merepresentasikan fitur kategorikal umum dalam permasalahan Machine Learning.

## 1. Import Library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Membuat Dataset Contoh

```python
# Membuat dataset dengan fitur kategorik
data = {
    'Warna': ['Merah', 'Biru', 'Hijau', 'Merah', 'Biru', 'Hijau', 'Merah', 'Biru'],
    'Ukuran': ['Kecil', 'Sedang', 'Besar', 'Sedang', 'Besar', 'Kecil', 'Besar', 'Sedang'],
    'Kualitas': ['Rendah', 'Sedang', 'Tinggi', 'Sedang', 'Tinggi', 'Rendah', 'Tinggi', 'Sedang'],
    'Harga': [100, 200, 300, 150, 350, 120, 320, 210]
}

df = pd.DataFrame(data)
print("Dataset Awal:")
display(df)
```

## 3. Label Encoding

**Label Encoding** mengubah setiap kategori unik menjadi angka integer. Cocok untuk kolom yang tidak memiliki urutan khusus (namun perlu hati-hati karena model bisa mengasumsikan ada urutan).

```python
# Label Encoding untuk kolom 'Warna'
df_label = df.copy()
le = LabelEncoder()

df_label['Warna_Encoded'] = le.fit_transform(df_label['Warna'])

print("Mapping Label Encoding untuk kolom 'Warna':")
for i, kelas in enumerate(le.classes_):
    print(f"  {kelas} -> {i}")

print("\nDataFrame setelah Label Encoding:")
display(df_label[['Warna', 'Warna_Encoded']])
```

## 4. Ordinal Encoding

**Ordinal Encoding** digunakan ketika data kategorik memiliki **urutan yang bermakna** (seperti Rendah < Sedang < Tinggi). Kita mendefinisikan urutan sendiri agar model mengetahui hierarki kategori.

```python
# Ordinal Encoding untuk kolom 'Ukuran' (Kecil < Sedang < Besar)
df_ordinal = df.copy()

oe_ukuran = OrdinalEncoder(categories=[['Kecil', 'Sedang', 'Besar']])
df_ordinal['Ukuran_Encoded'] = oe_ukuran.fit_transform(df_ordinal[['Ukuran']])

print("Mapping Ordinal Encoding untuk kolom 'Ukuran':")
print("  Kecil  -> 0")
print("  Sedang -> 1")
print("  Besar  -> 2")

print("\nDataFrame setelah Ordinal Encoding:")
display(df_ordinal[['Ukuran', 'Ukuran_Encoded']])
```

```python
# Ordinal Encoding untuk kolom 'Kualitas' (Rendah < Sedang < Tinggi)
oe_kualitas = OrdinalEncoder(categories=[['Rendah', 'Sedang', 'Tinggi']])
df_ordinal['Kualitas_Encoded'] = oe_kualitas.fit_transform(df_ordinal[['Kualitas']])

print("Mapping Ordinal Encoding untuk kolom 'Kualitas':")
print("  Rendah -> 0")
print("  Sedang -> 1")
print("  Tinggi -> 2")

print("\nDataFrame setelah Ordinal Encoding (Ukuran & Kualitas):")
display(df_ordinal[['Ukuran', 'Ukuran_Encoded', 'Kualitas', 'Kualitas_Encoded']])
```

## 5. One-Hot Encoding dengan pandas.get_dummies()

**One-Hot Encoding** membuat kolom baru (dummy variable) untuk setiap nilai unik dalam kolom kategorik. Ini adalah cara yang paling aman untuk menghindari asumsi urutan yang tidak perlu.

```python
# One-Hot Encoding untuk kolom 'Warna'
df_ohe = pd.get_dummies(df, columns=['Warna'], prefix='Warna')

print("DataFrame setelah One-Hot Encoding untuk kolom 'Warna':")
display(df_ohe)
```

## 6. Menggabungkan Semua Encoding

Berikut adalah hasil encoding lengkap dari semua kolom kategorik dalam satu DataFrame:

```python
# Encoding lengkap
df_final = df.copy()

# Label Encoding untuk 'Warna'
df_final['Warna_Label'] = LabelEncoder().fit_transform(df_final['Warna'])

# Ordinal Encoding untuk 'Ukuran'
oe = OrdinalEncoder(categories=[['Kecil', 'Sedang', 'Besar']])
df_final['Ukuran_Ordinal'] = oe.fit_transform(df_final[['Ukuran']])

# Ordinal Encoding untuk 'Kualitas'
oe2 = OrdinalEncoder(categories=[['Rendah', 'Sedang', 'Tinggi']])
df_final['Kualitas_Ordinal'] = oe2.fit_transform(df_final[['Kualitas']])

print("DataFrame dengan semua encoding:")
display(df_final)
```

## 7. Visualisasi Distribusi Data Sebelum dan Sesudah Encoding

```python
# Visualisasi distribusi kolom 'Warna' sebelum dan sesudah encoding
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Sebelum Encoding
df['Warna'].value_counts().plot(kind='bar', ax=axes[0], color=['#e74c3c', '#3498db', '#2ecc71'])
axes[0].set_title('Distribusi Warna (Sebelum Encoding)', fontsize=12)
axes[0].set_xlabel('Warna')
axes[0].set_ylabel('Frekuensi')
axes[0].tick_params(axis='x', rotation=0)

# Sesudah Encoding
df_final['Warna_Label'].value_counts().sort_index().plot(kind='bar', ax=axes[1], color=['#3498db', '#2ecc71', '#e74c3c'])
axes[1].set_title('Distribusi Warna (Setelah Label Encoding)', fontsize=12)
axes[1].set_xlabel('Kode Warna (0=Biru, 1=Hijau, 2=Merah)')
axes[1].set_ylabel('Frekuensi')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

print("\nKesimpulan:")
print("- Label Encoding mengubah kategori menjadi angka (0, 1, 2, ...) secara alfabetis.")
print("- Ordinal Encoding menggunakan urutan yang kita tentukan sendiri.")
print("- One-Hot Encoding membuat kolom biner untuk setiap kategori unik.")
```
