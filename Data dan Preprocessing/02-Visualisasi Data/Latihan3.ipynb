# Latihan 3: Visualisasi Data Exploratory Data Analysis (EDA)

Pada tahap ini, dilakukan proses visualisasi data menggunakan library **Seaborn** dan **Matplotlib** untuk menganalisis dataset `automobileEDA.csv`. Fokus utama dari latihan ini adalah:

## 1. Distribution Analysis
Mengamati sebaran harga mobil menggunakan **Histogram** dan **Kernel Density Estimation (KDE)** untuk menentukan rentang harga pasar.

## 2. Correlation Analysis
Mengidentifikasi hubungan linear antara fitur teknis (seperti `engine-size` dan `horsepower`) terhadap harga menggunakan **Regression Plot** dan **Heatmap**.

## 3. Categorical Comparison
Membandingkan variabel kategori seperti gaya bodi (`body-style`) dan jenis penggerak roda (`drive-wheels`) melalui **Box Plot** dan **Pie Chart**.

## 4. Multivariate Exploration
Menelaah interaksi kompleks antara beberapa variabel sekaligus untuk memahami faktor utama penentu nilai ekonomi kendaraan.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Memuat Dataset
df = pd.read_csv('automobileEDA.csv')

# Mengatur gaya visualisasi
sns.set(style="whitegrid")

# ==========================================================
# A. ANALISIS DISTRIBUSI (Histogram & KDE)
# ==========================================================
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], kde=True, color='blue')
plt.title('Distribusi Harga Mobil')
plt.show()

# ==========================================================
# B. ANALISIS HUBUNGAN LINEAR (Regression Plot)
# ==========================================================
plt.figure(figsize=(10, 5))
sns.regplot(x='engine-size', y='price', data=df)
plt.title('Ukuran Mesin vs Harga')
plt.show()

# ==========================================================
# C. ANALISIS KATEGORIKAL (Box Plot)
# ==========================================================
plt.figure(figsize=(10, 6))
sns.boxplot(x='body-style', y='price', data=df)
plt.title('Harga berdasarkan Gaya Bodi')
plt.show()

# ==========================================================
# D. ANALISIS FREKUENSI (Bar Chart / Count Plot)
# ==========================================================
plt.figure(figsize=(12, 6))
sns.countplot(x='make', data=df, order=df['make'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Jumlah Mobil per Merek')
plt.show()

# ==========================================================
# E. PIE CHART (Persentase Jenis Penggerak Roda)
# ==========================================================
plt.figure(figsize=(8, 8))
data_pie = df['drive-wheels'].value_counts()
plt.pie(data_pie, labels=data_pie.index, autopct='%1.1f%%', 
        startangle=140, colors=sns.color_palette('pastel'))
plt.title('Persentase Jenis Penggerak Roda (Drive Wheels)')
plt.show()

# ==========================================================
# F. ANALISIS KORELASI (Heatmap)
# ==========================================================
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()

# ==========================================================
# G. ANALISIS MULTIVARIAT (Scatter Plot dengan Hue)
# ==========================================================
plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='price', hue='drive-wheels', s=100, data=df)
plt.title('Tenaga Kuda vs Harga berdasarkan Drive Wheels')
plt.show()
```
