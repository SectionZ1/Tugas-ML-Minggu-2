### Keterangan Latihan:
Latihan ini bertujuan untuk mempraktikkan berbagai metode **Transformasi Data** yang mencakup:
1. **Imputasi Data Numerik** - Mean, Median, dan Arbitrary Imputation.
2. **Binning (Pemberitaan/Pengelompokan)** - Pengelompokan data kontinu menggunakan `pd.cut()` dan `pd.qcut()`.
3. **Penskalaan Data (Feature Scaling)** - Normalisasi (MinMaxScaler) dan Standardisasi (StandardScaler).

Dataset yang digunakan: Synthetic data, `bins.csv`, dan `scalling.csv`.

## 1. Import Library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

## 2. Imputasi Data
Kita akan membuat dataset buatan yang mengandung missing values (`NaN`) dan mempraktikkan cara mengatasinya.

```python
# Membuat dataset contoh
data_impute = {
    'umur': [29, 43, np.nan, 25, 34, np.nan, 50],
    'skor': [85, np.nan, 90, 78, np.nan, 95, 88]
}
df_impute = pd.DataFrame(data_impute)
print("DataFrame Awal dengan Missing Values:")
display(df_impute)

# A. Imputasi Mean
df_mean = df_impute.copy()
df_mean['umur'] = df_mean['umur'].fillna(df_mean['umur'].mean())
print("\nHasil Imputasi Mean pada Kolom 'umur':")
display(df_mean)

# B. Imputasi Median
df_median = df_impute.copy()
df_median['skor'] = df_median['skor'].fillna(df_median['skor'].median())
print("\nHasil Imputasi Median pada Kolom 'skor':")
display(df_median)

# C. Imputasi Arbitrary (Nilai Suka-suka)
df_arbitrary = df_impute.copy()
df_arbitrary['umur'] = df_arbitrary['umur'].fillna(99) # Mengganti NaN dengan 99
print("\nHasil Imputasi Arbitrary (99) pada Kolom 'umur':")
display(df_arbitrary)
```

## 3. Binning Data
Mengelompokkan data harga yang kontinu ke dalam 3 kategori ('Murah', 'Standar', 'Mahal') menggunakan dataset `bins.csv`.

```python
# Membaca dataset bins.csv
df_bins = pd.read_csv('bins.csv')
print("Dataset bins.csv Awal:")
display(df_bins)

# A. Binning berdasarkan interval yang sama (linspace)
bins_range = np.linspace(min(df_bins['Harga']), max(df_bins['Harga']), 4)
kategori = ['Murah', 'Standar', 'Mahal']
df_bins['Kategori_Harga_Cut'] = pd.cut(df_bins['Harga'], bins=bins_range, labels=kategori, include_lowest=True)

# B. Binning berdasarkan kuantil/distribusi frekuensi yang sama (qcut)
df_bins['Kategori_Harga_Qcut'] = pd.qcut(df_bins['Harga'], 3, labels=kategori)

print("\nHasil Binning dengan pd.cut dan pd.qcut:")
display(df_bins)
```

## 4. Scaling (Penskalaan Data)
Kita memuat data dari `scalling.csv` dan menerapkan Normalisasi (Min-Max Scaling) serta Standardisasi (Standard Scaling).

```python
# Membaca dataset scalling.csv
df_scale = pd.read_csv('scalling.csv')
print("Dataset scalling.csv Awal:")
display(df_scale)

# A. Normalisasi (Min-Max Scaling) -> rentang [0, 1]
min_max_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(
    min_max_scaler.fit_transform(df_scale), 
    columns=df_scale.columns
)
print("\nHasil Normalisasi (MinMaxScaler):")
display(df_minmax)

# B. Standardisasi (Standard Scaling) -> Mean = 0, Std Dev = 1
std_scaler = StandardScaler()
df_std = pd.DataFrame(
    std_scaler.fit_transform(df_scale), 
    columns=df_scale.columns
)
print("\nHasil Standardisasi (StandardScaler):")
display(df_std)
```

## 5. Kesimpulan
- **Imputasi** membantu kita mengisi missing values sehingga algoritma model dapat memproses baris data tersebut tanpa membuangnya.
- **Binning** mengubah fitur numerik kontinu menjadi kategorik ordinal untuk mempermudah pemahaman pola (seperti segmentasi umur atau harga).
- **Normalisasi (MinMaxScaler)** menskalakan data ke rentang 0-1 (bagus untuk algoritma berbasis jarak seperti KNN atau Neural Network).
- **Standardisasi (StandardScaler)** mereduksi deviasi sehingga berpusat pada nilai mean = 0 dan variansi = 1 (bagus untuk algoritma regresi linear, logistik, SVM, dll).
