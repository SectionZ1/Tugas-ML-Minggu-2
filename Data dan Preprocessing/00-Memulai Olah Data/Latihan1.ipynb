Saya menerapkan teori korelasi pada Probabilitas dan Statistika

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Penerapan Pandas 
df = pd.read_csv('Tab.csv', index_col=0) 
df['Kepadatan'] = df['Populasi'] / df["Area"]

# 2. Penerapan Numpy
df['Log_Populasi'] = np.round(np.log(df['Populasi']), 2)

# 3. Penerapan Scipy
korelasi, p_value = stats.pearsonr(df['Populasi'], df['Area'])

# 4. Penerapan Matplotlib
plt.figure(figsize=(10,5))
plt.bar(df['Negara'], df['Populasi'], color='red')
plt.title('Populasi Negara (Data dari Tab.csv)')
plt.xlabel('Negara')
plt.ylabel('Populasi')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Menampilkan Insight
print("== Insight Data ==")
print(df[['Negara', 'Kepadatan', 'Log_Populasi']])
print(f"\nKorelasi antara Populasi dan Area: {korelasi:.2f}")

if korelasi > 0.5:
    print("Kesimpulan: Terdapat korelasi positif yang kuat antara Populasi dan Area.")
else:
    print("Kesimpulan: Tidak terdapat korelasi yang kuat antara Populasi dan Area.")
```
