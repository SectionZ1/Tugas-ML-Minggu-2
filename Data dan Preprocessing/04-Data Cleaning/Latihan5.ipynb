```python
import pandas as pd
import numpy as np

print("="*50)
print("1. DATA CLEANING: DATASET BUKU (Flickr)")
print("="*50)
try:
    # Membaca data buku
    df_books = pd.read_csv('BL-Flickr-Images-Book.csv')
    
    # Membuang (drop) kolom yang tidak diperlukan
    to_drop = ['Edition Statement', 'Corporate Author', 'Corporate Contributors', 
               'Former owner', 'Engraver', 'Contributors', 'Issuance type', 'Shelfmarks']
    df_books.drop(columns=to_drop, inplace=True, errors='ignore')
    
    # Mengubah index menggunakan kolom 'Identifier'
    df_books.set_index('Identifier', inplace=True)
    
    # Membersihkan kolom 'Date of Publication' (Hanya mengambil 4 digit tahun)
    extr = df_books['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
    df_books['Date of Publication'] = pd.to_numeric(extr)
    
    print("Tampilan 3 Data Teratas (Dataset Buku setelah dibersihkan):")
    display(df_books.head(3))
except FileNotFoundError:
    print("Error: File 'BL-Flickr-Images-Book.csv' tidak ditemukan!")


print("\n" + "="*50)
print("2. DATA CLEANING: DATASET OLYMPICS")
print("="*50)
try:
    # Membaca data olympics (skip baris pertama karena merupakan header kotor)
    df_olympics = pd.read_csv('olympics.csv', header=1)
    
    # Mengganti nama kolom yang berantakan
    new_names =  {'Unnamed: 0': 'Country',
                  '? Summer': 'Summer Olympics',
                  '01 !': 'Gold_Summer',
                  '02 !': 'Silver_Summer',
                  '03 !': 'Bronze_Summer',
                  '? Winter': 'Winter Olympics',
                  '01 !.1': 'Gold_Winter',
                  '02 !.1': 'Silver_Winter',
                  '03 !.1': 'Bronze_Winter',
                  '? Games': 'Total Games',
                  '01 !.2': 'Gold_Total',
                  '02 !.2': 'Silver_Total',
                  '03 !.2': 'Bronze_Total'}
    df_olympics.rename(columns=new_names, inplace=True)
    
    # Mengatur kolom 'Country' sebagai indeks
    df_olympics.set_index('Country', inplace=True)
    
    print("Tampilan 3 Data Teratas (Dataset Olypmics setelah rename kolom):")
    display(df_olympics.head(3))
except FileNotFoundError:
    print("Error: File 'olympics.csv' tidak ditemukan!")


print("\n" + "="*50)
print("3. DATA CLEANING: DATASET UNIVERSITY TOWNS (Teks Berantakan)")
print("="*50)
try:
    # Membaca file txt yang formatnya berantakan baris demi baris
    towns = []
    with open('university_towns.txt') as file:
        for line in file:
            if '[edit]' in line:
                # Ingat State jika ada kata [edit]
                state = line.strip()
            else:
                # Selain itu, anggap sebagai RegionName (nama kota/universitas)
                towns.append([state, line.strip()])
                
    # Menjadikan DataFrame
    df_towns = pd.DataFrame(towns, columns=['State', 'RegionName'])
    
    # Membersihkan string: menghilangkan '[edit]' di State dan teks di dalam tanda kurung '( )' di RegionName
    df_towns['State'] = df_towns['State'].str.replace(r'\[edit\]', '', regex=True)
    df_towns['RegionName'] = df_towns['RegionName'].str.replace(r'\s\(.*', '', regex=True)
    
    print("Tampilan 5 Data Teratas (Dataset University Towns setelah dibersihkan):")
    display(df_towns.head())
except FileNotFoundError:
    print("Error: File 'university_towns.txt' tidak ditemukan!")
```
